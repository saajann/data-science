```python
import pandas as pd
import seaborn as sns
```

```python
players = pd.read_csv('../../data/football_players.csv')
```

```python
players.head()
```

```python
pd.set_option('display.max_columns', None)
players
```

```python
sns.scatterplot(data=players, x='age', y='value_eur')
```

```python
features = ['age', 'value_eur', 'height_cm', 'weight_kg', 'overall', 'skill_moves', 'pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic', 'wage_eur']
```

```python
X = players[features]
y = players['wage_eur']
```

```python
sns.heatmap(X.corr(numeric_only=True))
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X = X.drop(columns=['wage_eur'], axis=1)

X = scaler.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=200)

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, y_pred)
```
