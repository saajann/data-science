# SEABORN

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/car_crashes.csv')
```

```python
df.head()
```

## DISTRIBUTION PLOTS

### Distribution plot
Mostra la distrubuzione di una feature

```python
sns.displot(df['not_distracted'])
```

### Joint plot
Mette in confronto due distibuzioni e traccia uno scatter plot

```python
sns.jointplot(data=df, x='speeding', y='alcohol', kind='reg')
```
