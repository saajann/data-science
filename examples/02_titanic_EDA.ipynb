Importo le librerie necessarie per il progetto e carico il dataset dei passeggeri del Titanic

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/titanic.csv')
```

Dato che ci sono dei passeggeri che non hanno l'età specificata, li rimuovo dal mio DataFrame

```python
df.dropna(subset='Age', inplace=True)
df
```

Pulisco ulteriormente i dati in modo da non avere valori mancanti

```python
df.fillna(df['Fare'].mean(), inplace=True)
df.drop(columns='Cabin', inplace=True)
df.info()
```

```python
df
```

Esplorazione delle variabili categoriche e visualizzazione delle distribuzioni in grafici a barre

```python
survived = df['Survived'].value_counts()
sex = df['Sex'].value_counts()
pclass = df['Pclass'].value_counts()

plt.subplot(1,3,1)
plt.bar(survived.index, survived.values)

plt.subplot(1,3,2)
plt.bar(sex.index, sex.values)

plt.subplot(1,3,3)
plt.bar(pclass.index, pclass.values)

plt.show()
```

Cerco correlazioni tra variabili numeriche e creo una heatmap 

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
numeric_df.corr()

sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap delle correlazioni')
plt.show()
```

Cerco relazioni tra variabili categoriche 

```python
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Sopravvivenza in base alla Classe')
plt.show()
```
