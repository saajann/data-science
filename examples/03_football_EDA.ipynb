Esercitazione per consolidare le conoscenze di Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/football_players.csv')
```

## Distribution Plots

### Distribution Plot

```python
sns.displot(df['height_cm'])
```

### Histogram

```python
sns.histplot(df['age'])
```

### Joint Plot
Compara le distribuzioni di due variabili numeriche

```python
sns.jointplot(data=df, x='height_cm', y='weight_kg', kind='reg')
```

### KDE Plot

```python
sns.kdeplot(df['overall'])
```

### Pair Plot

```python
sns.pairplot(df)
```
