# EDA del dataset di giocatori di calcio
Esercitazione per consolidare le conoscenze di Seaborn, utilizzando un dataset di calciatori

## Importazione

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/football_players.csv')
```

## Distribution Plots

### Distribution Plot

```python
sns.displot(df['height_cm'])
```

### Histogram

```python
sns.histplot(df['age'])
```

### Joint Plot
Compara le distribuzioni di due variabili numeriche

```python
sns.jointplot(data=df, x='height_cm', y='weight_kg', kind='reg')
```

### KDE Plot

```python
sns.kdeplot(df['overall'])
```

### Pair Plot

```python
# sns.pairplot(df)
```

## Categorical Plots

### Bar Plot

```python
sns.barplot(data=df, x='preferred_foot', y='value_eur')
```

Si può notare che in media chi ha come piede principale il sinistro ha un valore di mercato più alto

### Count Plot

```python
sns.countplot(data=df, x='preferred_foot')
```

Noto che la maggior parte dei calciatori nel DataFrame ha come piede principale il destro

### Box Plot

```python
sns.boxplot(data=df[df['age'] <= 30], x='age', y='height_cm', hue='preferred_foot')
```

## Matrix Plots

### Heatmap

```python
columns = ['age', 'height_cm', 'weight_kg', 'overall', 'value_eur']
short_df = df[columns].loc[0:100]
sns.heatmap(short_df.corr(), annot=True)
```

Esempio semplice ma mi permette di capire che c'è una correlazione tra l'altezza e il peso dei giocatori
