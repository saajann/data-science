```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/football_players.csv')
```

```python
plt.subplot(1,2,1)
plt.plot(df['age'], df['overall'], marker='o', color='r', linestyle='')
plt.subplot(1,2,2)
plt.plot(df['age'], df['wage_eur'], marker='o', color='r', linestyle='')
plt.show()
```

```python
plt.plot(df['overall'], df['wage_eur'], marker='o', color='r', linestyle='')
plt.grid()
plt.show()
```

```python
plt.scatter(df['overall'], df['age'], cmap='viridis', alpha=0.1, s=100)
plt.show()
```
