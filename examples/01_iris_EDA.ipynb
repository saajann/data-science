Importo i moduli necessari, in questo caso Pandas e Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Carico il dataset Iris in un DataFrame Pandas

```python
df = pd.read_csv('../data/Iris.csv')
```

Stampo le prime 5 righe, per dare uno primo sguardo al mio DataFrame

```python
df.head()
```

Guardo quali colonne ci sono, di che tipo sono i dati e se ci sono celle vuote

```python
df.info()
```

Analizzo media / deviazione standard / min / max

```python
df.describe()
```

Per ogni specie conto quanti valori ci sono

```python
df['Species'].value_counts()
```

Traccio un grafico a barre con Matplotlib delle speci di Iris

```python
species_counts = df['Species'].value_counts()
plt.bar(species_counts.index, species_counts.values)
plt.show()
```

Traccio un istogramma della lunghezza del sepali

```python
plt.hist(df['SepalLengthCm'])
plt.show()
```

Scatter plot che mette in relazione la lunghezza con la larghezza del sepalo

```python
plt.scatter(df['SepalLengthCm'], df['SepalWidthCm'])
plt.show()
```

Con Seaborn posso differenziare le varie specie

```python
import seaborn as sns

sns.scatterplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species')
plt.show()
```

4 istogrammi, uno per ogni feature del fiore

```python
plt.subplot(2,2,1)
plt.hist(df['PetalLengthCm'])
plt.title('Petal Length')

plt.subplot(2,2,2)
plt.hist(df['PetalWidthCm'])
plt.title('Petal Width')

plt.subplot(2,2,3)
plt.hist(df['SepalLengthCm'])
plt.title('Sepal Length')

plt.subplot(2,2,4)
plt.hist(df['SepalWidthCm'])
plt.title('Sepal Width')

plt.show()
```
